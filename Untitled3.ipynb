### hashage et chiffrement avec haslib

```python
import pandas as pd
from cryptography.fernet import Fernet
import hashlib

# Générer une clé de chiffrement symétrique (à stocker de manière sécurisée)
key = Fernet.generate_key()
cipher_suite = Fernet(key)

# Fonction pour anonymiser un champ avec chiffrement symétrique
def anonymize_symmetric(data):
    return cipher_suite.encrypt(data.encode()).decode()

# Fonction pour désanonymiser un champ avec chiffrement symétrique
def deanonymize_symmetric(data):
    return cipher_suite.decrypt(data.encode()).decode()

# Fonction pour hasher un champ (pour garantir la cohérence des jointures)
def anonymize_hash(data):
    return hashlib.sha256(data.encode()).hexdigest()[:10]

# Exemple de données
data = {
    "nc": ["12345", "67890", "12345", "67890", "11111"],
    "id_cl": ["ABC123", "DEF456", "ABC123", "XYZ789", "XYZ789"],
    "nom_cl": ["Alice", "Bob", "Alice", "Charlie", "Charlie"],
    "code_ag": ["AGT1", "AGT2", "AGT1", "AGT3", "AGT3"],
    "nom_ag": ["Agent1", "Agent2", "Agent1", "Agent3", "Agent3"],
}
df = pd.DataFrame(data)

# Créer une table de correspondance pour id_cl
mapping_table = {}
for id_cl in df["id_cl"].unique():
    mapping_table[id_cl] = anonymize_hash(id_cl)

# Anonymisation des données
df_anonymized = df.copy()

# Anonymiser les champs sensibles avec chiffrement symétrique
df_anonymized["nc"] = df_anonymized["nc"].apply(anonymize_symmetric)
df_anonymized["nom_cl"] = df_anonymized["nom_cl"].apply(anonymize_symmetric)
df_anonymized["code_ag"] = df_anonymized["code_ag"].apply(anonymize_symmetric)
df_anonymized["nom_ag"] = df_anonymized["nom_ag"].apply(anonymize_symmetric)

# Anonymiser l'id_cl avec hachage en utilisant la table de correspondance
df_anonymized["id_cl"] = df_anonymized["id_cl"].map(mapping_table)


# Désanonymisation des données
df_deanonymized = df_anonymized.copy()

# Désanonymiser les champs sensibles avec chiffrement symétrique
df_deanonymized["nc"] = df_deanonymized["nc"].apply(deanonymize_symmetric)
df_deanonymized["nom_cl"] = df_deanonymized["nom_cl"].apply(deanonymize_symmetric)
df_deanonymized["code_ag"] = df_deanonymized["code_ag"].apply(deanonymize_symmetric)
df_deanonymized["nom_ag"] = df_deanonymized["nom_ag"].apply(deanonymize_symmetric)

# Désanonymiser l'id_cl en utilisant la table de correspondance
reverse_mapping_table = {v: k for k, v in mapping_table.items()}  # Inverser la table de correspondance
df_deanonymized["id_cl"] = df_deanonymized["id_cl"].map(reverse_mapping_table)

```

```python
hashlib :
Peu importe le fichier dans lequel la valeur se trouve, tant que:
la méthode de hachage utilisée reste la même (fonction SHA-256 /hmac + clé éventuelle), 
le hachage sera identique pour une même entrée.

1. Clé de chiffrement (pour le chiffrement symétrique) :
Rôle : La clé de chiffrement est utilisée pour chiffrer et déchiffrer les données sensibles de manière réversible.

Utilisation :

Chiffrement : Les colonnes sensibles (numero_compte, nom_client, code_agent, nom_agent) sont chiffrées avec un algorithme symétrique (comme AES ou Fernet).

Déchiffrement : Les mêmes colonnes peuvent être déchiffrées ultérieurement pour retrouver les valeurs en clair.

Stockage : La clé doit être stockée de manière sécurisée (par exemple, dans un gestionnaire de secrets comme AWS Secrets Manager, Azure Key Vault, ou un fichier chiffré).

Comportement : Le chiffrement symétrique génère des données différentes à chaque exécution, même si l'entrée est la même.

Pourquoi ? :

Les algorithmes de chiffrement symétrique (comme AES ou Fernet) utilisent un vecteur d'initialisation (IV) pour randomiser le processus de chiffrement. Cela garantit que deux chiffrements de la même donnée produisent des résultats différents, ce qui renforce la sécurité.

Par exemple, si vous chiffrez 12345 deux fois, vous obtiendrez deux résultats différents comme gAAAAABl... et hBBBBBCm....


Critère		SHA-256 (sans clé)	HMAC-SHA256 (avec clé)
Cohérence	Oui	     		Oui (si la clé est la même)
Réversibilité	Non			Non
Gestion de clé	Non nécessaire		Nécessaire
Protection rainbow tables Non		Oui
Complexité	Simple			Plus complexe (gestion de clé)
Cas d'utilisation	Anonymisation, vérification d'intégrité	 / Anonymisation sécurisée, authentification


2. Table de correspondance (pour le hachage) :
Rôle : La table de correspondance est utilisée pour mapper les valeurs originales des id_client à leurs versions hachées.

Utilisation :

Hachage : Les id_client sont hachés (par exemple, avec SHA-256) pour garantir que la même valeur aura toujours le même hachage, ce qui facilite les jointures entre fichiers.

Désanonymisation : La table de correspondance permet de retrouver la valeur originale d'un id_client à partir de son hachage.

Stockage : La table de correspondance doit également être stockée de manière sécurisée, car elle permet de retrouver les valeurs originales.


Fichiers chiffrés :

Stockez les clés dans un fichier chiffré sur un système de fichiers sécurisé.
```
